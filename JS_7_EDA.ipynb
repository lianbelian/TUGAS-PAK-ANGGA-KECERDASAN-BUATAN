# Pertanyaan Bisnis

1. Bagaimana pengaruh musim terhadap jumlah penyewaan sepeda?
2. Pada bulan apa jumlah penyewaan sepeda paling tinggi?

1

```python
season_rent = df.groupby('season')['cnt'].mean()
season_rent
```

```python
season_rent.plot(kind='bar')
plt.title('Rata-rata Penyewaan Berdasarkan Musim')
plt.xlabel('Season')
plt.ylabel('Jumlah Penyewaan')
plt.show()
```

2

```python
month_rent = df.groupby('mnth')['cnt'].mean()
month_rent
```

```python
month_rent.plot(kind='line', marker='o')
plt.title('Rata-rata Penyewaan Berdasarkan Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penyewaan')
plt.show()
```

Insight:

- Musim tertentu memiliki rata-rata penyewaan lebih tinggi dibanding musim lainnya.
- Bulan tertentu menunjukkan jumlah penyewaan tertinggi sehingga dapat menjadi periode puncak penggunaan sepeda.

Kesimpulan

1. Musim berpengaruh terhadap jumlah penyewaan sepeda.
2. Terdapat bulan dengan tingkat penyewaan tertinggi dibanding bulan lainnya.
3. Analisis data membantu menemukan pola penggunaan sepeda berdasarkan waktu tertentu.
