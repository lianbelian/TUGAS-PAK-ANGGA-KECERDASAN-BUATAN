3. **Menggunakan Fungsi MATH**

```python
import math

# 1. Trigonometri (dalam radian)
print(math.sin(1))
print(math.cos(1))
print(math.tan(1))

# 2. Logaritma
print(math.log(10))     # Logaritma natural
print(math.log10(100))  # Logaritma basis 10

# 3. Konversi & Jarak
print(math.degrees(1.57))      # Konversi radian ke derajat
print(math.dist((0,0), (3,4))) # Jarak antara dua titik

# 4. Teori Bilangan & Statistik
print(math.factorial(5))   # Faktorial (5!)
print(math.fmod(10, 3))     # Sisa bagi
print(math.isnan(0.5))      # Cek apakah bukan angka (NaN)
print(math.gamma(5))        # Fungsi Gamma
print(math.perm(5, 2))      # Permutasi
```

**TUGAS PRAKTIKUM**

```python
import math

# =========================
# 1. PERKALIAN MATRIKS
# =========================

A = [[3, 1],
     [2, 0]]

B = [[2, 1],
     [1, -1]]

C = [[-1, 1, 2],
     [2, 0, -1]]

def multiply_matrix(X, Y):
    result = [[0 for _ in range(len(Y[0]))] for _ in range(len(X))]
    for i in range(len(X)):
        for j in range(len(Y[0])):
            for k in range(len(Y)):
                result[i][j] += X[i][k] * Y[k][j]
    return result

print("1. HASIL PERKALIAN MATRIKS")
print("A x B =")
for row in multiply_matrix(A, B):
    print(row)

print("\nB x C =")
for row in multiply_matrix(B, C):
    print(row)


# =========================
# 2. FAKTORIAL
# =========================

def faktorial(n):
    hasil = 1
    for i in range(1, n+1):
        hasil *= i
    return hasil

print("\n2. FAKTORIAL")
print("6! =", faktorial(6))
print("8! / 5! =", faktorial(8) // faktorial(5))


# =========================
# 3. VEKTOR
# =========================

A_vec = [5, 3, 7]
B_vec = [12, -3, 1]

# Panjang vektor
def panjang(v):
    return math.sqrt(sum(x**2 for x in v))

# Dot product
def dot(v1, v2):
    return sum(a*b for a, b in zip(v1, v2))

# Cross product
def cross(v1, v2):
    return [
        v1[1]*v2[2] - v1[2]*v2[1],
        v1[2]*v2[0] - v1[0]*v2[2],
        v1[0]*v2[1] - v1[1]*v2[0]
    ]

print("\n3. VEKTOR")
print("Panjang A =", panjang(A_vec))
print("Panjang B =", panjang(B_vec))
print("A · B =", dot(A_vec, B_vec))
print("A x B =", cross(A_vec, B_vec))
```
