**1.	Central Tendency**

 a.	Mendapatkan Nilai Mean

```python
import numpy as np

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = np.mean(score)
print(x)
```

b.	Mendapatkan Nilai Median

```python
import numpy as np

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = np.median(score)
print(x)
```

c.	Mendapatkan Nilai Modus / Mode

```python
from scipy import stats

score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = stats.mode(score, keepdims=True)
print(x)
```

**2.	Spread**

a.	Mendapatkan Nilai Range

```python
import numpy as np

data = [2, 3, 7, 8, 9, 12, 15, 20, 50, 42]

minimum = np.min(data)
maximum = np.max(data)
x = maximum - minimum
print(x)
```

b.	Mendapatkan Nilai Percentile

```python
import numpy as np

data = [90, 80, 70, 65, 110, 80, 83, 67, 94]

x = np.percentile(data, 90)
print(x)
```

c.	Mendapatkan Nilai Quartile dan Interquartile Range

```python
import numpy as np

data = [3, 4, 4, 5, 5, 6, 6, 7, 8, 8, 9, 10]

Q1 = np.median(data[:int(len(data)/2)])
Q3 = np.median(data[int(len(data)/2):])
IQR = Q3 - Q1

print(Q1, Q3, IQR)
```

d.	Mendapatkan Nilai Variance dan Standart Deviation

```python
import numpy as np

data= [86, 87, 88, 86, 87, 85, 86]

mean = np.mean(data)
var = np.var(data)
std = np.std(data)

print(mean, var, std)
```

```python
data = [1,2,3,4,5,5,6,2,3,4,5,6,6,4,3,2,1,4,3,5,6,6,5,4,5]

# hitung frekuensi
frekuensi = {}

for angka in data:
    if angka in frekuensi:
        frekuensi[angka] += 1
    else:
        frekuensi[angka] = 1

# cari modus
modus = max(frekuensi, key=frekuensi.get)

print("Data:", data)
print("Frekuensi:", frekuensi)
print("Modus:", modus)
print("Jumlah kemunculan:", frekuensi[modus])
```

```python
import numpy as np

data = [50, 54, 62, 50, 52, 59, 61, 63, 65, 10, 53, 63, 65, 50, 59, 62, 50, 51, 57, 60, 63, 65, 65, 53, 99]

# a. Urutkan data
sorted_data = np.sort(data)
print("Data urut:", sorted_data)

# b. Q1, Q3, IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)

# c. Batas bawah & atas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

# d. Outlier
outlier = [x for x in data if x < lower_bound or x > upper_bound]

print("Outlier:", outlier)
```

```python
import numpy as np

# titik tengah
mid = np.array([43, 48, 52.5, 58, 63, 68])

# frekuensi
freq = np.array([2, 3, 6, 4, 5, 2])

# jumlah data
n = np.sum(freq)

# mean
mean = np.sum(mid * freq) / n

# varians
var = np.sum(freq * (mid - mean)**2) / n

# standar deviasi
std = np.sqrt(var)

print("Mean:", mean)
print("Varians:", var)
print("Standar Deviasi:", std)
```
